```python
def repetir(n):
    def decorador(func):
        def envoltura(*args, **kwargs):
            for _ in range(n):
                func(*args, **kwargs)
        return envoltura
    return decorador


@repetir(3)
def saludar(nombre):
    print(f"Hola, {nombre}!")


saludar("Carlos")
```

```python
from functools import wraps

def mi_decorador(func):
    @wraps(func)
    def envoltura(*args, **kwargs):
        print(" Antes de ejecutar la funci贸n")
        resultado = func(*args, **kwargs)
        print(" Despu茅s de ejecutar la funci贸n")
        return resultado
    return envoltura


@mi_decorador
def saludar(nombre):
    print(f"Hola, {nombre}!")

saludar("Carlos")


```
