# Polimorfismo

```python
class Gato:
    def sonido(self):
        return "Miau"

class Perro:
    def sonido(self):
        return "Guau"

def hacer_sonar(animal):
    print(animal.sonido())

gato = Gato()
perro = Perro()

hacer_sonar(gato)
hacer_sonar(perro)
```
